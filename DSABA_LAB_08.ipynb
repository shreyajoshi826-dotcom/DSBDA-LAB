```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

dataset = pd.read_csv('titanic.csv')  # your downloaded file
print(dataset.head())
```

```
print(dataset.shape)     # should show (891, columns)
print(dataset.info())    # data types
print(dataset.describe())  # summary statistics
```

```
sns.countplot(x='Sex', data=dataset)
plt.title("Gender Count")
plt.show()
```

```
sns.countplot(x='Survived', data=dataset)
plt.title("Survival Count")
plt.show()
```

```
sns.countplot(x='Sex', hue='Survived', data=dataset)
plt.title("Survival based on Gender")
plt.show()
```

```
sns.histplot(dataset['Age'], bins=10, kde=True)
plt.title("Age Distribution")
plt.show()
```

```
sns.jointplot(x='Age', y='Fare', data=dataset, kind='scatter')
```

```
sns.boxplot(x='Sex', y='Age', data=dataset)
plt.show()
```

```
sns.violinplot(x='Sex', y='Age', data=dataset, hue='Survived')
plt.show()
```

```
sns.swarmplot(x='Sex', y='Age', data=dataset, hue='Survived')
plt.show()
```

```
corr = dataset.corr(numeric_only=True)
sns.heatmap(corr, annot=True)
plt.show()
```

```
sns.histplot(dataset['Fare'], bins=10, kde=False)
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Number of Passengers")
plt.show()
```
